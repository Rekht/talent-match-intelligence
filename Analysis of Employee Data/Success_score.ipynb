```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
import matplotlib.pyplot as plt
import seaborn as sns

class RobustOptimalWeightFinder:
    def __init__(self):
        self.optimal_weights = {}
    
    def find_optimal_weights(self, df, target_col='is_high'):
        """
        Mencari weights optimal - SUPER ROBUST terhadap data kosong
        """
        print(" Mencari Optimal Weights...")
        
        # Handle target variable dengan robust
        if target_col not in df.columns:
            print(" Target column tidak ditemukan, menggunakan equal weights")
            return self._get_equal_weights(['cognitive', 'behavioral', 'competency', 'contextual'])
        
        # Filter hanya rows dengan target yang valid
        valid_mask = ~df[target_col].isna()
        if valid_mask.sum() == 0:
            print(" Tidak ada data target yang valid, menggunakan equal weights")
            return self._get_equal_weights(['cognitive', 'behavioral', 'competency', 'contextual'])
        
        df_valid = df[valid_mask].copy()
        y = df_valid[target_col]
        
        # 1. Hitung component scores dengan handling extreme robustness
        component_scores = self._calculate_component_scores_robust(df_valid)
        
        if not component_scores:
            print(" Tidak ada component scores yang bisa dihitung")
            return self._get_equal_weights(['cognitive', 'behavioral', 'competency', 'contextual'])
        
        # 2. Gunakan metode yang paling robust
        weights = self._most_robust_weight_method(component_scores, y)
        
        self.optimal_weights = weights
        return weights
    
    def _calculate_component_scores_robust(self, df):
        """
        Menghitung component scores dengan extreme robustness
        """
        components = {}
        
        # 1. COGNITIVE SCORE - handle semua kemungkinan error
        cognitive_score = self._safe_calculate_cognitive(df)
        if cognitive_score is not None:
            components['cognitive'] = cognitive_score
        
        # 2. BEHAVIORAL SCORE
        behavioral_score = self._safe_calculate_behavioral(df)
        if behavioral_score is not None:
            components['behavioral'] = behavioral_score
        
        # 3. COMPETENCY SCORE
        competency_score = self._safe_calculate_competency(df)
        if competency_score is not None:
            components['competency'] = competency_score
        
        # 4. CONTEXTUAL SCORE
        contextual_score = self._safe_calculate_contextual(df)
        if contextual_score is not None:
            components['contextual'] = contextual_score
        
        return components
    
    def _safe_calculate_cognitive(self, df):
        """Calculate cognitive score dengan extreme safety"""
        try:
            cognitive_features = ['pauli', 'faxtor', 'iq', 'gtq', 'tiki']
            available_features = [f for f in cognitive_features if f in df.columns]
            
            if not available_features:
                return None
            
            scores = []
            for feature in available_features:
                if feature in df.columns:
                    col_data = df[feature]
                    # Handle NaN dan infinite values
                    clean_data = col_data.replace([np.inf, -np.inf], np.nan)
                    if clean_data.isna().all():
                        continue
                    
                    # Fill NaN dengan median
                    median_val = np.nanmedian(clean_data)
                    if np.isnan(median_val):
                        continue
                    
                    filled_data = clean_data.fillna(median_val)
                    
                    # Normalize
                    if feature == 'tiki':  # Reverse scoring
                        max_val = filled_data.max()
                        if max_val > 0:
                            score = 1 - (filled_data / max_val)
                        else:
                            score = filled_data * 0  # zero array
                    else:
                        max_val = filled_data.max()
                        if max_val > 0:
                            score = filled_data / max_val
                        else:
                            score = filled_data * 0  # zero array
                    
                    scores.append(score)
            
            if scores:
                # Convert to numpy array untuk handling yang aman
                scores_array = np.array(scores)
                # Mean along axis 0, handle all-NaN slices
                with warnings.catch_warnings():
                    warnings.simplefilter("ignore", category=RuntimeWarning)
                    final_score = np.nanmean(scores_array, axis=0)
                    # Replace remaining NaN dengan 0.5 (neutral)
                    final_score = np.where(np.isnan(final_score), 0.5, final_score)
                return final_score
            
        except Exception as e:
            print(f"  Cognitive score error: {e}")
        
        return None
    
    def _safe_calculate_behavioral(self, df):
        """Calculate behavioral score dengan extreme safety"""
        try:
            scores = []
            
            # DISC Scoring
            if 'disc' in df.columns:
                disc_map = {'D': 0.9, 'DC': 0.85, 'DI': 0.8, 'CD': 0.8, 'I': 0.7, 
                           'ID': 0.75, 'C': 0.8, 'CS': 0.75, 'S': 0.6, 'SC': 0.65}
                disc_score = df['disc'].map(lambda x: disc_map.get(x, 0.5))
                # Handle NaN
                disc_score = disc_score.fillna(0.5)
                scores.append(disc_score.values)
            
            # MBTI Scoring  
            if 'mbti' in df.columns:
                mbti_map = {'ENTJ': 0.9, 'INTJ': 0.85, 'ESTJ': 0.8, 'ENTP': 0.8,
                           'INTP': 0.75, 'ENFJ': 0.7, 'INFJ': 0.7, 'ESTP': 0.65}
                mbti_score = df['mbti'].map(lambda x: mbti_map.get(x, 0.5))
                mbti_score = mbti_score.fillna(0.5)
                scores.append(mbti_score.values)
            
            # PAPI Traits - hanya jika ada data
            papi_cols = [col for col in df.columns if col.startswith('Papi_')]
            if papi_cols:
                papi_data = df[papi_cols].copy()
                # Fill NaN dengan median per column
                for col in papi_cols:
                    if col in papi_data.columns:
                        median_val = papi_data[col].median()
                        if np.isnan(median_val):
                            papi_data[col] = 5.0  # default neutral
                        else:
                            papi_data[col] = papi_data[col].fillna(median_val)
                
                papi_score = papi_data.mean(axis=1) / 9.0
                scores.append(papi_score.values)
            
            if scores:
                scores_array = np.array(scores)
                with warnings.catch_warnings():
                    warnings.simplefilter("ignore", category=RuntimeWarning)
                    final_score = np.nanmean(scores_array, axis=0)
                    final_score = np.where(np.isnan(final_score), 0.5, final_score)
                return final_score
                
        except Exception as e:
            print(f"  Behavioral score error: {e}")
        
        return None
    
    def _safe_calculate_competency(self, df):
        """Calculate competency score dengan extreme safety"""
        try:
            competency_cols = ['CEX', 'CSI', 'FTC', 'GDR', 'IDS', 'LIE', 'QDD', 'SEA', 'STO', 'VCU']
            available_cols = [col for col in competency_cols if col in df.columns]
            
            if not available_cols:
                return None
            
            competency_data = df[available_cols].copy()
            
            # Fill NaN dengan median per column
            for col in available_cols:
                if col in competency_data.columns:
                    median_val = competency_data[col].median()
                    if np.isnan(median_val):
                        competency_data[col] = 3.0  # default neutral (scale 1-5)
                    else:
                        competency_data[col] = competency_data[col].fillna(median_val)
            
            competency_score = competency_data.mean(axis=1) / 5.0
            return competency_score.values
            
        except Exception as e:
            print(f"  Competency score error: {e}")
        
        return None
    
    def _safe_calculate_contextual(self, df):
        """Calculate contextual score dengan extreme safety"""
        try:
            scores = []
            
            # Experience
            if 'years_of_service_months' in df.columns:
                exp_data = df['years_of_service_months'].copy()
                exp_data = exp_data.fillna(exp_data.median())
                if len(exp_data) > 0:
                    exp_score = 1 - abs(exp_data - 60) / 120
                    exp_score = np.clip(exp_score, 0, 1)
                    scores.append(exp_score.values)
            
            # Education
            if 'education_id' in df.columns:
                edu_data = df['education_id'].copy()
                edu_data = edu_data.fillna(edu_data.median())
                if len(edu_data) > 0:
                    max_edu = edu_data.max()
                    if max_edu > 0:
                        edu_score = edu_data / max_edu
                        scores.append(edu_score.values)
            
            # Grade level
            if 'grade_id' in df.columns:
                grade_data = df['grade_id'].copy()
                grade_data = grade_data.fillna(grade_data.median())
                if len(grade_data) > 0:
                    max_grade = grade_data.max()
                    if max_grade > 0:
                        grade_score = grade_data / max_grade
                        scores.append(grade_score.values)
            
            if scores:
                scores_array = np.array(scores)
                with warnings.catch_warnings():
                    warnings.simplefilter("ignore", category=RuntimeWarning)
                    final_score = np.nanmean(scores_array, axis=0)
                    final_score = np.where(np.isnan(final_score), 0.5, final_score)
                return final_score
                
        except Exception as e:
            print(f"  Contextual score error: {e}")
        
        return None
    
    def _most_robust_weight_method(self, component_scores, y):
        """
        Metode paling robust untuk mencari weights
        """
        try:
            # Convert to DataFrame untuk memudahkan processing
            X_df = pd.DataFrame(component_scores)
            
            # Handle any remaining NaN
            X_df = X_df.fillna(0.5)
            
            # Pastikan y tidak ada NaN
            y_clean = y.fillna(0) if hasattr(y, 'fillna') else y
            
            # Method 1: Correlation-based (paling robust)
            weights_corr = self._correlation_weights(X_df, y_clean)
            
            # Method 2: Variance-based (jika correlation gagal)
            weights_var = self._variance_weights(X_df)
            
            # Combine methods
            final_weights = {}
            for component in component_scores.keys():
                w1 = weights_corr.get(component, 0.25)
                w2 = weights_var.get(component, 0.25)
                # Prioritize correlation weights, fallback to variance
                final_weights[component] = w1 if w1 > 0 else w2
            
            # Normalize
            total = sum(final_weights.values())
            if total > 0:
                final_weights = {k: v/total for k, v in final_weights.items()}
            else:
                final_weights = self._get_equal_weights(component_scores.keys())
            
            return final_weights
            
        except Exception as e:
            print(f"  Weight calculation error: {e}, using equal weights")
            return self._get_equal_weights(component_scores.keys())
    
    def _correlation_weights(self, X_df, y):
        """Weight berdasarkan correlation - sangat robust"""
        weights = {}
        
        for col in X_df.columns:
            try:
                x_vals = X_df[col].values
                y_vals = y.values if hasattr(y, 'values') else y
                
                # Pastikan tidak ada NaN
                mask = ~(np.isnan(x_vals) | np.isnan(y_vals))
                if mask.sum() < 2:  # butuh minimal 2 points untuk correlation
                    weights[col] = 0
                    continue
                
                x_clean = x_vals[mask]
                y_clean = y_vals[mask]
                
                if len(np.unique(x_clean)) < 2 or len(np.unique(y_clean)) < 2:
                    weights[col] = 0
                    continue
                
                corr = np.corrcoef(x_clean, y_clean)[0, 1]
                weights[col] = abs(corr) if not np.isnan(corr) else 0
                
            except Exception:
                weights[col] = 0
        
        return weights
    
    def _variance_weights(self, X_df):
        """Weight berdasarkan variance - fallback method"""
        weights = {}
        
        for col in X_df.columns:
            try:
                variance = np.nanvar(X_df[col].values)
                weights[col] = variance if not np.isnan(variance) else 0
            except Exception:
                weights[col] = 0
        
        return weights
    
    def _get_equal_weights(self, components):
        """Return equal weights sebagai final fallback"""
        if isinstance(components, dict):
            components = list(components.keys())
        n = len(components)
        return {comp: 1.0/n for comp in components}
    
    def calculate_success_score(self, df, weights=None):
        """
        Calculate final success score dengan weights optimal
        """
        if weights is None:
            weights = self.optimal_weights
        
        component_scores = self._calculate_component_scores_robust(df)
        
        if not component_scores:
            # Return neutral score jika tidak ada component
            return np.array([0.5] * len(df))
        
        success_scores = np.zeros(len(df))
        for component, weight in weights.items():
            if component in component_scores:
                success_scores += component_scores[component] * weight
        
        return success_scores
    
    def plot_optimal_weights(self, year=None):
        """Visualisasi optimal weights"""
        if not self.optimal_weights:
            print("No optimal weights found.")
            return
        
        plt.figure(figsize=(10, 6))
        components = list(self.optimal_weights.keys())
        weights = [self.optimal_weights[comp] for comp in components]
        
        colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4']
        bars = plt.bar(components, weights, color=colors[:len(components)], alpha=0.8)
        
        title = 'Optimal Weights untuk Success Score'
        if year:
            title += f' - Tahun {year}'
        plt.title(title, fontsize=16, fontweight='bold')
        plt.ylabel('Weight Percentage', fontsize=12)
        
        # Add percentage labels
        for bar, weight in zip(bars, weights):
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{weight:.1%}', ha='center', va='bottom', fontweight='bold')
        
        plt.grid(axis='y', alpha=0.3)
        plt.tight_layout()
        plt.show()

# MAIN EXECUTION YANG SUPER ROBUST
def analyze_success_patterns_all_years(all_dataframes):
    """
    Analisis success patterns untuk semua tahun - SUPER ROBUST VERSION
    """
    yearly_results = {}
    
    for year, df in all_dataframes.items():
        print(f"\n{'='*60}")
        print(f" ANALYZING YEAR {year}")
        print(f"{'='*60}")
        
        # Basic data info
        print(f" Data Shape: {df.shape}")
        
        # Check target variable
        if 'is_high' not in df.columns:
            print(" 'is_high' column tidak ditemukan, skip tahun ini")
            continue
        
        high_performers = df['is_high'].sum() if 'is_high' in df.columns else 0
        total_employees = len(df)
        print(f" High Performers: {high_performers} / {total_employees} ({high_performers/total_employees:.1%})")
        
        # Initialize weight finder
        weight_finder = RobustOptimalWeightFinder()
        
        try:
            # Find optimal weights
            optimal_weights = weight_finder.find_optimal_weights(df)
            
            # Calculate success scores
            success_scores = weight_finder.calculate_success_score(df, optimal_weights)
            
            # Store results
            yearly_results[year] = {
                'optimal_weights': optimal_weights,
                'success_scores': success_scores,
                'weight_finder': weight_finder
            }
            
            # Print results
            print(f" OPTIMAL WEIGHTS TAHUN {year}:")
            for component, weight in optimal_weights.items():
                print(f"    {component.upper():<12}: {weight:.1%}")
            
            # Calculate correlation jika memungkinkan
            try:
                correlation = np.corrcoef(success_scores, df['is_high'])[0, 1]
                print(f" Correlation dengan High Performance: {correlation:.3f}")
            except:
                print(" Correlation: Cannot calculate")
            
            # Plot weights
            weight_finder.plot_optimal_weights(year)
            
        except Exception as e:
            print(f" Error analyzing year {year}: {str(e)[:100]}...")
            print(" Skip ke tahun berikutnya...")
            continue
    
    return yearly_results

# IMPORT WARNINGS
import warnings
warnings.filterwarnings('ignore')

# JALANKAN ANALISIS
print(" MEMULAI ANALISIS SUKSES PATTERN 2021-2025")
print("VERSION: SUPER ROBUST - HANDLES ALL MISSING DATA")

# Asumsikan all_dataframes sudah terdefinisi
yearly_results = analyze_success_patterns_all_years({
    2021: all_data_2021,
    2022: all_data_2022, 
    2023: all_data_2023,
    2024: all_data_2024,
    2025: all_data_2025
})

# SUMMARY RESULTS
if yearly_results:
    print(f"\n{'='*60}")
    print(" FINAL SUMMARY ALL YEARS")
    print(f"{'='*60}")
    
    summary_data = {}
    for year, results in yearly_results.items():
        summary_data[year] = results['optimal_weights']
    
    summary_df = pd.DataFrame(summary_data).fillna(0)
    print(summary_df.round(3))
    
    # Plot trends
    plt.figure(figsize=(12, 6))
    for component in summary_df.index:
        plt.plot(summary_df.columns, summary_df.loc[component], 
                marker='o', linewidth=2, label=component, markersize=6)
    
    plt.title('Evolution of Success Score Weights (2021-2025)', fontsize=14, fontweight='bold')
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Weight', fontsize=12)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()

else:
    print("\n Tidak ada hasil yang berhasil dihitung untuk semua tahun")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pointbiserialr
import warnings
warnings.filterwarnings('ignore')

class SuccessPatternAnalyzer:
    def __init__(self):
        self.results = {}
        
    def comprehensive_analysis(self, df, year):
        """
        Analisis komprehensif pattern success untuk satu tahun dengan grouping proper
        """
        print(f"\n COMPREHENSIVE ANALYSIS YEAR {year}")
        print("=" * 60)
        
        # 1. DATA EXPLORATION
        self._explore_data(df, year)
        
        # 2. COMPREHENSIVE FEATURE ANALYSIS  
        feature_analysis = self._comprehensive_feature_analysis(df, year)
        
        # 3. COMPONENT SCORING DENGAN GROUPING YANG BENAR
        components = self._calculate_components_proper(df)
        
        # 4. WEIGHT OPTIMIZATION
        optimal_weights = self._optimize_weights(components, df['is_high'])
        
        # 5. SUCCESS SCORE CALCULATION
        success_scores = self._calculate_success_score(components, optimal_weights)
        
        # 6. VALIDATION & EVALUATION
        validation_results = self._validate_success_score(success_scores, df['is_high'])
        
        # Store results
        self.results[year] = {
            'optimal_weights': optimal_weights,
            'success_scores': success_scores,
            'validation': validation_results,
            'feature_analysis': feature_analysis,
            'components': components
        }
        
        return self.results[year]
    
    def _explore_data(self, df, year):
        """Exploratory Data Analysis"""
        print(f" DATA OVERVIEW - {year}")
        print(f"   Total Employees: {len(df):,}")
        print(f"   High Performers: {df['is_high'].sum():,} ({df['is_high'].mean():.1%})")
        
        # Check data completeness
        numeric_cols = df.select_dtypes(include=[np.number]).columns
        completeness = df[numeric_cols].notna().mean().mean()
        print(f"   Data Completeness: {completeness:.1%}")
    
    def _comprehensive_feature_analysis(self, df, year):
        """
        Comprehensive feature analysis dengan grouping yang proper
        """
        print(f"\n COMPREHENSIVE FEATURE ANALYSIS - {year}")
        
        categories_summary = {}
        
        # 1. COMPETENCY PILLARS (Performance Metrics)
        print(" 1. COMPETENCY PILLARS:")
        competency_features = ['CEX', 'CSI', 'FTC', 'GDR', 'IDS', 'LIE', 'QDD', 'SEA', 'STO', 'VCU']
        competency_correlations = {}
        
        for feature in competency_features:
            if feature in df.columns:
                clean_data = df[feature].fillna(3.0)
                corr = np.corrcoef(clean_data, df['is_high'])[0, 1]
                competency_correlations[feature] = corr
                print(f"   {feature}: {corr:7.3f}")
        
        if competency_correlations:
            avg_competency_corr = np.mean(list(competency_correlations.values()))
            categories_summary['competency'] = avg_competency_corr
            print(f"    AVERAGE: {avg_competency_corr:.3f}")
        
        # 2. PSYCHOMETRIC PROFILES (Cognitive + Personality)
        print(f"\n 2. PSYCHOMETRIC PROFILES:")
        
        # 2A. Cognitive Abilities
        print("    A. Cognitive Abilities:")
        cognitive_features = ['pauli', 'faxtor', 'iq', 'gtq', 'tiki']
        cognitive_correlations = {}
        
        for feature in cognitive_features:
            if feature in df.columns:
                clean_data = df[feature].fillna(df[feature].median())
                corr = np.corrcoef(clean_data, df['is_high'])[0, 1]
                cognitive_correlations[feature] = corr
                print(f"      {feature:8}: {corr:7.3f}")
        
        # 2B. PAPI Personality Traits
        print("    B. PAPI Personality Traits:")
        papi_features = [f'Papi_{chr(i)}' for i in range(65, 91)]  # A to Z
        papi_correlations = {}
        
        papi_available = [f for f in papi_features if f in df.columns]
        for feature in papi_available[:8]:  # Tampilkan 8 pertama
            clean_data = df[feature].fillna(5.0)
            corr = np.corrcoef(clean_data, df['is_high'])[0, 1]
            papi_correlations[feature] = corr
            print(f"      {feature:8}: {corr:7.3f}")
        
        if len(papi_available) > 8:
            print(f"      ... and {len(papi_available) - 8} more PAPI traits")
        
        # 2C. Behavioral Profiles
        print("    C. Behavioral Profiles:")
        behavioral_correlations = {}
        
        if 'disc' in df.columns:
            disc_mapping = {'D': 0.9, 'DC': 0.85, 'DI': 0.8, 'CD': 0.8, 'I': 0.7, 
                           'ID': 0.75, 'C': 0.8, 'CS': 0.75, 'S': 0.6, 'SC': 0.65}
            disc_numeric = df['disc'].map(lambda x: disc_mapping.get(x, 0.5))
            corr_disc = np.corrcoef(disc_numeric.fillna(0.5), df['is_high'])[0, 1]
            behavioral_correlations['disc'] = corr_disc
            print(f"      {'disc':8}: {corr_disc:7.3f}")
        
        if 'mbti' in df.columns:
            mbti_mapping = {'ENTJ': 0.9, 'INTJ': 0.85, 'ESTJ': 0.8, 'ENTP': 0.8,
                           'INTP': 0.75, 'ENFJ': 0.7, 'INFJ': 0.7, 'ESTP': 0.65}
            mbti_numeric = df['mbti'].map(lambda x: mbti_mapping.get(x, 0.5))
            corr_mbti = np.corrcoef(mbti_numeric.fillna(0.5), df['is_high'])[0, 1]
            behavioral_correlations['mbti'] = corr_mbti
            print(f"      {'mbti':8}: {corr_mbti:7.3f}")
        
        # Calculate overall psychometric correlation
        all_psychometric = (list(cognitive_correlations.values()) + 
                          list(papi_correlations.values()) + 
                          list(behavioral_correlations.values()))
        if all_psychometric:
            avg_psychometric_corr = np.mean(all_psychometric)
            categories_summary['psychometric'] = avg_psychometric_corr
            print(f"    AVERAGE: {avg_psychometric_corr:.3f}")
        
        # 3. STRENGTHS DATA (Gallup Strengths)
        print(f"\n 3. STRENGTHS DATA:")
        strength_features = [f'strength_rank_{i}' for i in range(1, 15)]
        
        # Strength categories mapping
        strength_categories = {
            'Strategic': 0.9, 'Analytical': 0.85, 'Input': 0.8, 'Intellection': 0.8,
            'Achiever': 0.85, 'Focus': 0.8, 'Responsibility': 0.75, 'Discipline': 0.75,
            'Command': 0.8, 'Activator': 0.75, 'Competition': 0.7, 'Self-Assurance': 0.7,
            'Relator': 0.6, 'Empathy': 0.6, 'Harmony': 0.55, 'Adaptability': 0.6,
            'Restorative': 0.7, 'Deliberative': 0.65, 'Context': 0.6, 'Futuristic': 0.7,
            'Ideation': 0.7, 'Learner': 0.65, 'Maximizer': 0.7, 'Developer': 0.65,
            'Connectedness': 0.6, 'Individualization': 0.65, 'Positivity': 0.6, 'Woo': 0.55,
            'Communication': 0.7, 'Includer': 0.6, 'Significance': 0.7, 'Arranger': 0.65,
            'Consistency': 0.6, 'Belief': 0.6
        }
        
        strength_correlations = []
        for i, strength_feature in enumerate(strength_features, 1):
            if strength_feature in df.columns:
                # Convert strength names ke numerical scores dengan weighting
                weight = (15 - i) / 14.0  # 1.0 untuk rank1, ~0.07 untuk rank14
                strength_numeric = df[strength_feature].map(lambda x: strength_categories.get(x, 0.5) * weight)
                corr = np.corrcoef(strength_numeric.fillna(0.3), df['is_high'])[0, 1]
                strength_correlations.append(corr)
        
        if strength_correlations:
            avg_strength_corr = np.mean(strength_correlations)
            categories_summary['strengths'] = avg_strength_corr
            print(f"    AVERAGE: {avg_strength_corr:.3f}")
            print(f"   Sample strengths: {[df[f'strength_rank_{i}'].iloc[0] for i in range(1, 4) if f'strength_rank_{i}' in df.columns]}")
        
        # 4. CONTEXTUAL FACTORS
        print(f"\n 4. CONTEXTUAL FACTORS:")
        contextual_correlations = []
        
        # Education
        if 'education_id' in df.columns:
            edu_corr = np.corrcoef(df['education_id'].fillna(df['education_id'].median()), df['is_high'])[0, 1]
            contextual_correlations.append(edu_corr)
            print(f"    Education:       {edu_corr:7.3f}")
        
        # Grade/Position Level  
        if 'grade_id' in df.columns:
            grade_corr = np.corrcoef(df['grade_id'].fillna(df['grade_id'].median()), df['is_high'])[0, 1]
            contextual_correlations.append(grade_corr)
            print(f"    Grade Level:     {grade_corr:7.3f}")
        
        # Experience
        if 'years_of_service_months' in df.columns:
            exp_corr = np.corrcoef(df['years_of_service_months'].fillna(df['years_of_service_months'].median()), df['is_high'])[0, 1]
            contextual_correlations.append(exp_corr)
            print(f"    Experience:      {exp_corr:7.3f}")
        
        if contextual_correlations:
            avg_contextual_corr = np.mean(contextual_correlations)
            categories_summary['contextual'] = avg_contextual_corr
            print(f"    AVERAGE: {avg_contextual_corr:.3f}")
        
        # SUMMARY PER CATEGORY
        print(f"\n CATEGORY SUMMARY - {year}:")
        for category, avg_corr in categories_summary.items():
            print(f"   {category.upper():12}: {avg_corr:7.3f}")
        
        return categories_summary
    
    def _calculate_components_proper(self, df):
        """
        Calculate components dengan grouping yang proper
        """
        components = {}
        
        # 1. COMPETENCY PILLARS
        competency_features = ['CEX', 'CSI', 'FTC', 'GDR', 'IDS', 'LIE', 'QDD', 'SEA', 'STO', 'VCU']
        available_competency = [f for f in competency_features if f in df.columns]
        
        if available_competency:
            competency_data = df[available_competency].fillna(3.0)
            components['competency'] = competency_data.mean(axis=1) / 5.0
        
        # 2. PSYCHOMETRIC PROFILES (Cognitive + Personality)
        psychometric_scores = []
        
        # 2A. Cognitive Abilities
        cognitive_features = ['pauli', 'faxtor', 'iq', 'gtq', 'tiki']
        available_cognitive = [f for f in cognitive_features if f in df.columns]
        
        if available_cognitive:
            cognitive_scores = []
            for feature in available_cognitive:
                if feature == 'tiki':
                    score = 1 - (df[feature] / df[feature].max())
                else:
                    score = df[feature] / df[feature].max()
                cognitive_scores.append(score.fillna(score.median()))
            
            psychometric_scores.append(np.mean(cognitive_scores, axis=0))
        
        # 2B. PAPI Personality Traits
        papi_features = [f'Papi_{chr(i)}' for i in range(65, 91)]
        available_papi = [f for f in papi_features if f in df.columns]
        
        if available_papi:
            papi_data = df[available_papi].fillna(5.0)
            papi_score = papi_data.mean(axis=1) / 9.0
            psychometric_scores.append(papi_score)
        
        # 2C. Behavioral Profiles
        behavioral_scores = []
        
        if 'disc' in df.columns:
            disc_mapping = {'D': 0.9, 'DC': 0.85, 'DI': 0.8, 'CD': 0.8, 'I': 0.7, 
                           'ID': 0.75, 'C': 0.8, 'CS': 0.75, 'S': 0.6, 'SC': 0.65}
            disc_score = df['disc'].map(lambda x: disc_mapping.get(x, 0.5))
            behavioral_scores.append(disc_score.fillna(0.5))
        
        if 'mbti' in df.columns:
            mbti_mapping = {'ENTJ': 0.9, 'INTJ': 0.85, 'ESTJ': 0.8, 'ENTP': 0.8,
                           'INTP': 0.75, 'ENFJ': 0.7, 'INFJ': 0.7, 'ESTP': 0.65}
            mbti_score = df['mbti'].map(lambda x: mbti_mapping.get(x, 0.5))
            behavioral_scores.append(mbti_score.fillna(0.5))
        
        if behavioral_scores:
            psychometric_scores.append(np.mean(behavioral_scores, axis=0))
        
        if psychometric_scores:
            components['psychometric'] = np.mean(psychometric_scores, axis=0)
        
        # 3. STRENGTHS DATA
        strength_features = [f'strength_rank_{i}' for i in range(1, 15)]
        available_strengths = [f for f in strength_features if f in df.columns]
        
        if available_strengths:
            strength_categories = {
                'Strategic': 0.9, 'Analytical': 0.85, 'Input': 0.8, 'Intellection': 0.8,
                'Achiever': 0.85, 'Focus': 0.8, 'Responsibility': 0.75, 'Discipline': 0.75,
                'Command': 0.8, 'Activator': 0.75, 'Competition': 0.7, 'Self-Assurance': 0.7,
                'Relator': 0.6, 'Empathy': 0.6, 'Harmony': 0.55, 'Adaptability': 0.6,
                'Restorative': 0.7, 'Deliberative': 0.65, 'Context': 0.6, 'Futuristic': 0.7,
                'Ideation': 0.7, 'Learner': 0.65, 'Maximizer': 0.7, 'Developer': 0.65,
                'Connectedness': 0.6, 'Individualization': 0.65, 'Positivity': 0.6, 'Woo': 0.55,
                'Communication': 0.7, 'Includer': 0.6, 'Significance': 0.7, 'Arranger': 0.65,
                'Consistency': 0.6, 'Belief': 0.6
            }
            
            strength_scores = []
            for i, strength_feature in enumerate(available_strengths, 1):
                weight = (15 - i) / 14.0
                strength_score = df[strength_feature].map(lambda x: strength_categories.get(x, 0.5) * weight)
                strength_scores.append(strength_score.fillna(0.3))
            
            if strength_scores:
                components['strengths'] = np.mean(strength_scores, axis=0)
        
        # 4. CONTEXTUAL FACTORS
        contextual_scores = []
        
        # Education
        if 'education_id' in df.columns:
            edu_score = df['education_id'] / df['education_id'].max()
            contextual_scores.append(edu_score.fillna(edu_score.median()))
        
        # Grade Level
        if 'grade_id' in df.columns:
            grade_score = df['grade_id'] / df['grade_id'].max()
            contextual_scores.append(grade_score.fillna(grade_score.median()))
        
        # Experience (optimal curve around 5 years/60 months)
        if 'years_of_service_months' in df.columns:
            exp_data = df['years_of_service_months'].fillna(60)
            exp_score = 1 - np.abs(exp_data - 60) / 120
            contextual_scores.append(np.clip(exp_score, 0, 1))
        
        if contextual_scores:
            components['contextual'] = np.mean(contextual_scores, axis=0)
        
        return components
    
    def _optimize_weights(self, components, ground_truth):
        """Optimize weights berdasarkan correlation dengan ground truth"""
        print(f"\n WEIGHT OPTIMIZATION")
        
        correlations = {}
        
        for component, scores in components.items():
            valid_mask = ~(np.isnan(scores) | np.isnan(ground_truth))
            if valid_mask.sum() > 10:
                clean_scores = scores[valid_mask]
                clean_truth = ground_truth[valid_mask]
                
                if len(np.unique(clean_scores)) > 1 and len(np.unique(clean_truth)) > 1:
                    corr = np.corrcoef(clean_scores, clean_truth)[0, 1]
                    correlations[component] = abs(corr) if not np.isnan(corr) else 0
                else:
                    correlations[component] = 0
            else:
                correlations[component] = 0
        
        if sum(correlations.values()) == 0:
            weights = {comp: 1/len(components) for comp in components.keys()}
        else:
            total_corr = sum(correlations.values())
            weights = {comp: corr/total_corr for comp, corr in correlations.items()}
        
        print("   Optimal Weights Found:")
        for component, weight in weights.items():
            print(f"    {component.upper():<12}: {weight:.1%} (corr: {correlations[component]:.3f})")
        
        return weights
    
    def _calculate_success_score(self, components, weights):
        """Calculate final success score"""
        success_score = np.zeros(len(next(iter(components.values()))))
        
        for component, weight in weights.items():
            if component in components:
                success_score += components[component] * weight
        
        return success_score
    
    def _validate_success_score(self, success_scores, ground_truth):
        """Validate success score against ground truth"""
        print(f"\n VALIDATION RESULTS")
        
        validation = {}
        
        # 1. Overall Correlation
        valid_mask = ~(np.isnan(success_scores) | np.isnan(ground_truth))
        if valid_mask.sum() > 10:
            clean_scores = success_scores[valid_mask]
            clean_truth = ground_truth[valid_mask]
            
            correlation = np.corrcoef(clean_scores, clean_truth)[0, 1]
            validation['correlation'] = correlation
            print(f"   Correlation with Ground Truth: {correlation:.3f}")
        
        # 2. Top-N Precision
        for k in [50, 100, 200]:
            if len(success_scores) >= k:
                top_k_indices = np.argsort(success_scores)[-k:]
                precision = ground_truth.iloc[top_k_indices].mean()
                validation[f'precision@{k}'] = precision
                print(f"   Precision@{k}: {precision:.1%}")
        
        # 3. Score Distribution Analysis
        high_performer_scores = success_scores[ground_truth == 1]
        other_scores = success_scores[ground_truth == 0]
        
        if len(high_performer_scores) > 0 and len(other_scores) > 0:
            score_gap = np.mean(high_performer_scores) - np.mean(other_scores)
            validation['score_gap'] = score_gap
            print(f"   Score Gap (High vs Others): {score_gap:.3f}")
        
        return validation
    
    def plot_comprehensive_analysis(self, year):
        """Plot comprehensive analysis untuk satu tahun"""
        if year not in self.results:
            print(f"No results for year {year}")
            return
        
        results = self.results[year]
        
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle(f'Success Pattern Analysis - Year {year}', fontsize=16, fontweight='bold')
        
        # 1. Weight Distribution
        weights = results['optimal_weights']
        axes[0,0].pie(weights.values(), labels=weights.keys(), autopct='%1.1f%%', startangle=90)
        axes[0,0].set_title('Optimal Weights Distribution')
        
        # 2. Score Distribution by Performance
        success_scores = results['success_scores']
        ground_truth = self.results[year]['validation']
        
        # Get actual ground truth from original data
        high_scores = success_scores[ground_truth == 1] if hasattr(ground_truth, '__len__') else []
        other_scores = success_scores[ground_truth == 0] if hasattr(ground_truth, '__len__') else []
        
        if len(high_scores) > 0 and len(other_scores) > 0:
            axes[0,1].hist([other_scores, high_scores], bins=20, alpha=0.7, 
                          label=['Other Performers', 'High Performers'], color=['blue', 'red'])
            axes[0,1].set_title('Success Score Distribution')
            axes[0,1].legend()
            axes[0,1].set_xlabel('Success Score')
            axes[0,1].set_ylabel('Frequency')
        
        # 3. Feature Importance dari comprehensive analysis
        if 'feature_analysis' in results:
            feature_importance = results['feature_analysis']
            axes[1,0].barh(list(feature_importance.keys()), list(feature_importance.values()))
            axes[1,0].set_title('Category Correlations with High Performance')
            axes[1,0].set_xlabel('Average Correlation')
        
        # 4. Component Scores Correlation
        components = results['components']
        component_corrs = {}
        
        for comp_name, comp_scores in components.items():
            valid_mask = ~(np.isnan(comp_scores) | np.isnan(ground_truth))
            if valid_mask.sum() > 10:
                corr = np.corrcoef(comp_scores[valid_mask], ground_truth[valid_mask])[0, 1]
                component_corrs[comp_name] = abs(corr)
        
        if component_corrs:
            axes[1,1].bar(component_corrs.keys(), component_corrs.values())
            axes[1,1].set_title('Component Correlation with High Performance')
            axes[1,1].set_ylabel('Absolute Correlation')
            axes[1,1].tick_params(axis='x', rotation=45)
        
        plt.tight_layout()
        plt.show()

# MAIN EXECUTION
def run_comprehensive_analysis(all_dataframes):
    """Run comprehensive analysis untuk semua tahun"""
    analyzer = SuccessPatternAnalyzer()
    
    yearly_summary = {}
    
    for year, df in all_dataframes.items():
        print(f"\n{'='*60}")
        print(f" YEAR {year} - COMPREHENSIVE ANALYSIS")
        print(f"{'='*60}")
        
        try:
            results = analyzer.comprehensive_analysis(df, year)
            yearly_summary[year] = results
            
            # Plot untuk tahun ini
            analyzer.plot_comprehensive_analysis(year)
            
        except Exception as e:
            print(f" Analysis failed for {year}: {str(e)[:100]}...")
            continue
    
    return analyzer, yearly_summary

# JALANKAN ANALISIS KOMPREHENSIF
print(" MEMULAI ANALISIS SUKSES PATTERN KOMPREHENSIF")
print("WITH PROPER GROUPING: Competency, Psychometric, Strengths, Contextual")

analyzer, results = run_comprehensive_analysis({
    2021: all_data_2021,
    2022: all_data_2022,
    2023: all_data_2023, 
    2024: all_data_2024,
    2025: all_data_2025
})

# FINAL SUMMARY ACROSS YEARS
if results:
    print(f"\n{'='*60}")
    print(" FINAL CROSS-YEAR SUMMARY")
    print(f"{'='*60}")
    
    summary_data = []
    for year, year_results in results.items():
        weights = year_results['optimal_weights']
        validation = year_results['validation']
        
        summary_data.append({
            'year': year,
            'competency_weight': weights.get('competency', 0),
            'psychometric_weight': weights.get('psychometric', 0),
            'strengths_weight': weights.get('strengths', 0),
            'contextual_weight': weights.get('contextual', 0),
            'correlation': validation.get('correlation', 0),
            'precision@100': validation.get('precision@100', 0)
        })
    
    summary_df = pd.DataFrame(summary_data)
    print(summary_df.round(3))
    
    # Calculate average weights across years
    print(f"\n AVERAGE WEIGHTS ACROSS ALL YEARS:")
    avg_weights = summary_df[['competency_weight', 'psychometric_weight', 'strengths_weight', 'contextual_weight']].mean()
    for category, weight in avg_weights.items():
        print(f"   {category.replace('_weight', '').upper():12}: {weight:.1%}")

else:
    print(" No results obtained from analysis")
```
