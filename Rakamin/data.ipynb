# Import

```python
import pandas as pd

# Read dataset
file_path = "Study Case DA.xlsx"
```

```python
# Menampilkan semua kolom tanpa terpotong
pd.set_option('display.max_columns', None)
```

# Read

## Employee

```python
# Sheet utama
employees = pd.read_excel(file_path, sheet_name="employees")
display(employees)
```

## Performance

```python
# Baca file Excel
performance = pd.read_excel(file_path, sheet_name="performance_yearly")

# Pivot data agar 'year' jadi kolom, dan 'rating' jadi nilai
performance_pivot = performance.pivot_table(
    index='employee_id',
    columns='year',
    values='rating'
).reset_index()

# (Opsional) urutkan kolom tahun
performance_pivot = performance_pivot[['employee_id'] + sorted([c for c in performance_pivot.columns if c != 'employee_id'])]

display(performance_pivot)
```

## Competencies

```python
# Baca data
competencies = pd.read_excel(file_path, sheet_name="competencies_yearly")

# Gabungkan pilar dan tahun jadi satu nama kolom
competencies['pillar_year'] = competencies['pillar_code'] + competencies['year'].astype(str)

# Pivot: setiap kolom = kombinasi pilar + tahun
competencies_wide = competencies.pivot_table(
    index='employee_id',
    columns='pillar_year',
    values='score'
).reset_index()

# Supaya kolom tidak multiindex
competencies_wide.columns.name = None

display(competencies_wide)
```

## Profiles Psych

```python
profiles = pd.read_excel(file_path, sheet_name="profiles_psych")
display(profiles)
```

## Papi Scores

```python
# Membaca data
papi = pd.read_excel(file_path, sheet_name="papi_scores")

# Membuat pivot: setiap employee_id jadi 1 baris, setiap scale_code jadi kolom
papi_pivot = papi.pivot_table(
    index="employee_id",
    columns="scale_code",
    values="score"
).reset_index()

# Urutkan kolom: employee_id dulu, lalu semua scale_code
papi_pivot = papi_pivot[['employee_id'] + sorted([c for c in papi_pivot.columns if c != 'employee_id'])]

# Tampilkan hasil
display(papi_pivot)
```

## Strengths

```python
# Baca file Excel
strength = pd.read_excel(file_path, sheet_name="strengths")

# Pivot data: jadikan rank sebagai kolom dan theme sebagai nilai
strength_pivot = strength.pivot_table(
    index="employee_id",
    columns="rank",
    values="theme",
    aggfunc="first"
).reset_index()

# Ganti nama kolom supaya jadi Rank1, Rank2, Rank3, dst
strength_pivot.columns = [
    "employee_id" if col == "employee_id" else f"Rank{int(col)}"
    for col in strength_pivot.columns
]

# Urutkan kolom Rank1, Rank2, Rank3, ...
rank_cols = sorted([c for c in strength_pivot.columns if c.startswith("Rank")],
                   key=lambda x: int(x.replace("Rank", "")))
strength_pivot = strength_pivot[["employee_id"] + rank_cols]

# Tampilkan hasil
display(strength_pivot)
```

## Dimension

```python
pillars = pd.read_excel(file_path, sheet_name="dim_competency_pillars")
# Dimensi organisasi
dim_grades = pd.read_excel(file_path, sheet_name="dim_grades")
dim_directorates = pd.read_excel(file_path, sheet_name="dim_directorates")
dim_divisions = pd.read_excel(file_path, sheet_name="dim_divisions")
dim_departments = pd.read_excel(file_path, sheet_name="dim_departments")
dim_positions = pd.read_excel(file_path, sheet_name="dim_positions")
dim_education = pd.read_excel(file_path, sheet_name="dim_education")
dim_majors = pd.read_excel(file_path, sheet_name="dim_majors")
dim_areas = pd.read_excel(file_path, sheet_name="dim_areas")
dim_companies = pd.read_excel(file_path, sheet_name="dim_companies")
```

```python
display(pillars, dim_grades, dim_directorates, dim_divisions, dim_departments,
        dim_positions, dim_education, dim_majors, dim_areas, dim_companies)
```

```python
# Buat dictionary mapping dari setiap tabel dimensi
map_company = dict(zip(dim_companies["company_id"], dim_companies["name"]))
map_area = dict(zip(dim_areas["area_id"], dim_areas["name"]))
map_position = dict(zip(dim_positions["position_id"], dim_positions["name"]))
map_department = dict(zip(dim_departments["department_id"], dim_departments["name"]))
map_division = dict(zip(dim_divisions["division_id"], dim_divisions["name"]))
map_directorate = dict(zip(dim_directorates["directorate_id"], dim_directorates["name"]))
map_grade = dict(zip(dim_grades["grade_id"], dim_grades["name"]))
map_education = dict(zip(dim_education["education_id"], dim_education["name"]))
map_major = dict(zip(dim_majors["major_id"], dim_majors["name"]))

# Replace ID dengan nama berdasarkan mapping
employees["company_id"] = employees["company_id"].map(map_company)
employees["area_id"] = employees["area_id"].map(map_area)
employees["position_id"] = employees["position_id"].map(map_position)
employees["department_id"] = employees["department_id"].map(map_department)
employees["division_id"] = employees["division_id"].map(map_division)
employees["directorate_id"] = employees["directorate_id"].map(map_directorate)
employees["grade_id"] = employees["grade_id"].map(map_grade)
employees["education_id"] = employees["education_id"].map(map_education)
employees["major_id"] = employees["major_id"].map(map_major)

# Ubah nama kolom agar lebih deskriptif
employees = employees.rename(columns={
    "company_id": "company_name",
    "area_id": "area_name",
    "position_id": "position_name",
    "department_id": "department_name",
    "division_id": "division_name",
    "directorate_id": "directorate_name",
    "grade_id": "grade_name",
    "education_id": "education_name",
    "major_id": "major_name"
})

display(employees)

```

```python

# 1️⃣ Mulai dari tabel employees (yang sudah pakai nama, bukan ID)
master = employees.copy()

# 2️⃣ Gabungkan Performance (pivot sudah berbentuk employee_id | 2021..2025)
master = master.merge(performance_pivot, on="employee_id", how="left")

# 3️⃣ Gabungkan Competencies (pivot sudah berbentuk employee_id | CEX2021..VCU2025)
master = master.merge(competencies_wide, on="employee_id", how="left")

# 4️⃣ Gabungkan Profiles Psych (IQ, GTQ, MBTI, dll)
master = master.merge(profiles, on="employee_id", how="left")

# 5️⃣ Gabungkan PAPI Scores (sudah dipivot jadi employee_id | Papi_A..Z)
master = master.merge(papi_pivot, on="employee_id", how="left")

# 6️⃣ Gabungkan Strengths (Rank1–Rank14)
master = master.merge(strength_pivot, on="employee_id", how="left")

# ======================================================
# 🔍 CEK HASIL GABUNGAN
# ======================================================
print("Jumlah baris master dataset:", len(master))
print("Jumlah kolom:", len(master.columns))
display(master)

# (Opsional) Simpan untuk analisis berikutnya
master.to_csv("master_dataset.csv", index=False)
print("\n✅ File 'master_dataset.csv' berhasil disimpan.")
```
