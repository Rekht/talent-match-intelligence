```python
# Import library
import pandas as pd

# Menampilkan semua kolom penuh
pd.set_option('display.max_columns', None)
```

```python
master = pd.read_csv("master_dataset.csv")
display(master.head())
```

```python
print(master.isna().sum().to_string())
```

```python
# Buat daftar tahun yang ingin difilter
years = [2021, 2022, 2023, 2024, 2025]

# Simpan hasil filter ke dictionary (biar rapi)
dfs = {}

for y in years:
    col = str(y)
    dfs[y] = master[master[col] == 5].copy()     # ambil baris dengan nilai 5
    print(f"✅ Tahun {y} → {len(dfs[y])} karyawan rating 5")
```
